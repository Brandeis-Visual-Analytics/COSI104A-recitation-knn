### Building a k-Nearest Neighbors Model

Below, you see two arrays containing data.  The first array, `training_data`, contains features (an x and a y value) and a `target` value, i.e. a label.  Your job is to write a k-Nearest Neighbors model that will use the kNN algorithm to generate a label for the `testing_data`.  It doesn't need to be fancy - just make the empty method do the right thing - assign the correct kNN label to those 5 items.

We have provided some skeleton code that will visualize your results.

Note that we hardcode the `k` value to be 4 here.

```python
training_data = [
    {"features": [4.0, 2.1], "target": "apple"},
    {"features": [5.0, 1.9], "target": "apple"},
    {"features": [10.0, 2.4], "target": "orange"},
    {"features": [4.0, 1.7], "target": "apple"},
    {"features": [3.0, 1.6], "target": "apple"},
    {"features": [8.0, 2.5], "target": "orange"},
    {"features": [14.0, 2.4], "target": "orange"},
    {"features": [18.0, 2.2], "target": "apple"},
    {"features": [5.0, 1.1], "target": "orange"},
    {"features": [12.0, 2.1], "target": "orange"}
  ]

testing_data = [
    {"features": [1.0, 2.7]},
    {"features": [11.2, 4.3]},
    {"features": [14.33, 8.1]},
    {"features": [-2.1,1.1 ]},
    {"features": [7.6, 1.9]},
]
```

```python
import matplotlib.pyplot as plt

def kNN_my_data(training_data, testing_data):
    # your code should assign a "target" of either "apple" or "orange" to each testing data item and return it.
    new_data = []
    for t in testing_data:
        # Here, you will want to add the correct target, according to the training data.
        new_data.append({"features": t["features"]})
    return new_data

```

```python
def visualize_knn_data(training_data, testing_data):
    apples_x = [item["features"][0] for item in training_data if ("target" in item) and item["target"] == "apple"]
    apples_y = [item["features"][1] for item in training_data if ("target" in item) and item["target"] == "apple"]
    oranges_x = [item["features"][0] for item in training_data if ("target" in item) and item["target"] == "orange"]
    oranges_y = [item["features"][1] for item in training_data if ("target" in item) and item["target"] == "orange"]

    plt.scatter(apples_x, apples_y, color='red', label='apple', s=100)
    plt.scatter(oranges_x, oranges_y, color='orange', label='orange', s=100)

    apples_testx = [item["features"][0] for item in testing_data if ("target" in item) and item["target"] == "apple"]
    apples_testy = [item["features"][1] for item in testing_data if ("target" in item) and item["target"] == "apple"]
    oranges_testx = [item["features"][0] for item in testing_data if ("target" in item) and item["target"] == "orange"]
    oranges_testy = [item["features"][1] for item in testing_data if ("target" in item) and item["target"] == "orange"]
    other_testx = [item["features"][0] for item in testing_data if not ('target' in item)]
    other_testy = [item["features"][1] for item in testing_data if not ('target' in item)]

    plt.scatter(apples_testx, apples_testy, color='red', label='apple', s=100, marker='x')
    plt.scatter(oranges_testx, oranges_testy, color='orange', label='orange', s=100, marker='x')
    plt.scatter(other_testx, other_testy, color='black', label='black', s=100, marker='x')

    plt.title("apples vs. oranges knn")
    plt.legend()

visualize_knn_data(training_data, testing_data)
```
